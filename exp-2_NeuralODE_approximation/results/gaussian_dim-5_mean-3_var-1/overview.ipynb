```python
import pandas as pd
import matplotlib.image as mpimg
import matplotlib.pyplot as plt
```

```python
fig, axes = plt.subplots(1, 3, figsize=(15, 5))
    
img1 = mpimg.imread('train_convergence_cov.png')
axes[0].imshow(img1)
axes[0].axis('off')

img2 = mpimg.imread('train_convergence_mean.png')
axes[1].imshow(img2)
axes[1].axis('off')

img3 = mpimg.imread('train_convergence_var.png')
axes[2].imshow(img3)
axes[2].axis('off')

plt.tight_layout()
plt.show()
```
